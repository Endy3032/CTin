```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
nptitanic = pd.read_csv("datasets/titanic.csv")
nptitanic.head()
```

Filtering Rows

```python
nptitanic.Pclass == 1
```

```python
nptitanic[nptitanic.Pclass == 1].head()
```

```python
nptitanic[nptitanic.Pclass == 1].head()
```

`isin()`

```python
nptitanic[nptitanic["Age"].isin([20, 21, 22])].head()
```

Multiple Filters

```python
nptitanic[(nptitanic["Age"].isin([20, 21, 22])) & (nptitanic["Pclass"] == 1)].head()
```

Filter Columns

```python
nptitanic.filter(["Name", "Sex", "Age"]).head()
```

Drop Columns

```python
nptitanic.drop(["Name", "Sex", "Age"], axis=1).head()
```
