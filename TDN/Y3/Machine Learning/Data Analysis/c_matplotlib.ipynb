## MatPlotLib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0, 20, 20)
b = np.sqrt(a)
plt.plot(a, b)
```

```python
fig = plt.figure()
ax = plt.axes()
ax.plot(a, b)
```

```python
del ax
del fig
```

```python
plt.rcParams["figure.figsize"] = (6, 3)
```

```python
plt.plot(a, b)
```

```python
plt.xlabel("X vals")
plt.ylabel("Y vals")
plt.title("Square root of X")
plt.plot(a, b)
```

```python
plt.plot(a, b, "r")
```

```python
plt.xlabel("X vals")
plt.ylabel("Y vals")
plt.title("Square root of X")
plt.plot(a, b, "r", label="square root")
plt.legend(loc="upper center")
```

```python
a = np.linspace(0, 10, 5)

plt.xlabel("X vals")
plt.ylabel("Y vals")
plt.plot(a, np.sqrt(a), "r", label="square root")
plt.plot(a, np.exp(a), "g", label="exp")
plt.legend(loc="upper center")
```

```python
pdiris = pd.read_csv("datasets/iris.csv")
pdiris.head()
```

```python
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.title("Sepal vs Petal Length")
plt.plot(pdiris["sepal_length"], pdiris["petal_length"], "g")
```

```python
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.title("Sepal vs Petal Length")
plt.scatter(pdiris["sepal_length"], pdiris["petal_length"], c="g")
```

```python
plt.pie(
	np.random.random_integers(100, 500, 4),
	explode=[0.05, 0.05, 0.05, 0.05],
	labels=["a", "b", "c", "d"],
	autopct="%1.1f%%",
)
plt.show()
```
