### 2.3

```python
abc = {
    "cname": [
        "Shyam&co",
        "Ralam&bros",
        "Sharma Enterprises",
        "Verma Furnitures",
        "Rahul store",
    ],
    "profit": [-10000, 10000, -5000, 15000, 20000],
}
df = pd.DataFrame(abc)
print(df)


def apply_function(profit, threshold=0):
    mean_profit = df["profit"].mean()
    return profit >= threshold * mean_profit


print("\nAfter apply function")

df["profit"] = df["profit"].apply(apply_function)
df
```
